```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
from scipy.stats import mannwhitneyu
import itertools

%matplotlib inline
```

```python
fn = "LFS_cleaned.tsv"
df = pd.read_csv(fn,sep="\t")
```

```python
df.nunique()
```

```python
df.corr()['Age_at_diagnosis']
```

## 1. The effect of gender

```python
gender_age = df.groupby('Sex')['Age_at_diagnosis'].mean()
for gender,age in gender_age.items():
    print('- Average "age at diagnosis" of {}: {:.1f}'.format(gender,age))
```

```python
plt.figure(figsize=(3,1))
ax = sns.barplot(x='Age_at_diagnosis',y='Sex', data=df,palette=['r','b'])
```

```python
# examine whether the distribution of age at diagnosis is Gaussian distribution
plt.figure(figsize=(10, 3))
for i, gender in enumerate(['F','M']):
    age_disr = df[df['Sex']==gender]['Age_at_diagnosis']
    plt.subplot(1,2,i+1)
    age_disr.plot.hist(bins=20,title=gender)
    _, p = stats.normaltest(age_disr)
    if p < 0.05:
        print("{} is not Gaussian distribution (the null hypothesis can be rejected, p={:.1e})".format(gender, p))
    else:
        print("{} is Gaussian distribution (the null hypothesis cannot be rejected, p={:.1e})".format(gender, p))
```

```python
_, p = mannwhitneyu(df[df['Sex']=='F']['Age_at_diagnosis'], 
                    df[df['Sex']=='M']['Age_at_diagnosis'])
print('Mann-Whitney test p-value = {:.3f}'.format(p))
```

__Conclusion (1)__
* Female has a significantly later age at diagnosis than male (29.4 yo > 28.5 yo, p =0.048)

## 2. The effect of tumor sites

```python
print("Number of distinct tumor sites: {}".format(len(set(df['Short_topo_combined'].tolist()))))
```

#### 2.1. Barplot of the number of pateints by tumor sites

```python
# Use seaborn to plot barplot based on the value counts of column "Short_topo_combined"
plt.figure(figsize=(10,30))
ax = sns.countplot(y='Short_topo_combined', data=df, order = df['Short_topo_combined'].value_counts().index)
```

#### 2.2. Simplify tumor sites by classifying the least common sites as 'Others'   

```python
# get the list of top 4 most frequent tumor sites
common_tumors = df['Short_topo_combined'].value_counts().index.tolist()[:4]
# convert least common tumor sites to "Others"
df['simplified_tumor'] = df.apply(lambda row: row['Short_topo_combined'] if row['Short_topo_combined'] in common_tumors else 'Others',axis=1)
```

```python
# Count the number of pateints by "simplified_tumor"
plt.figure(figsize=(6,2))
ax = sns.countplot(y='simplified_tumor', data=df, order = df['simplified_tumor'].value_counts().index)
```

#### 2.3. Barplot of 'age at diagnosis' by tumor sites

```python
plt.figure(figsize=(6,2))
ax = sns.barplot(x= 'Age_at_diagnosis', y='simplified_tumor', data=df)
```

```python
print("- Adrenal gland cancer has the earliest 'age at diagnosis': {:.1f} yo".format(
    np.mean(df[df['simplified_tumor'].isin(['ADRENAL GLAND'])]['Age_at_diagnosis'].tolist())))
print("- Breast cancer has the latest 'age at diagnosis': {:.1f} yo".format(
    np.mean(df[df['simplified_tumor'].isin(['BREAST'])]['Age_at_diagnosis'].tolist())))
```

#### 2.4. Compare the 'age at diagnosis' difference with Mann-Whitney test 

```python
def cal_age_diff(groups,column_name):
    
    list_pairs = list(itertools.combinations(groups, 2)) # get all possible pairs
    no_sig = []
    
    print('* Group with significant difference (Mann-Whitney test p-value < 0.05):')
    for pair in list_pairs:
        _, p = mannwhitneyu(df[df[column_name]==pair[0]]['Age_at_diagnosis'], 
                            df[df[column_name]==pair[1]]['Age_at_diagnosis'])
        if p<0.05:
            print('\t-{} vs {} (p-value = {:.1e})'.format(pair[0],pair[1],p))
        else:
            no_sig.append(" vs ".join(pair))
    
    print('* Group with no significant difference: {}'.format("; ".join(no_sig)))
    
```

```python
cal_age_diff(set(df['simplified_tumor'].tolist()),'simplified_tumor')
```

#### 2.5. Compare the age at diagnosis difference between male and female in each tumor site

```python
# Visualize the number of male and female in each tumor type
plt.figure(figsize=(6,3))
# Count average age of first onset by gender in each cancer
ax = sns.countplot(y='simplified_tumor', hue='Sex',palette=['r','b'], data=df)
```

```python
print('Gender distribution in breast cancer is unbalanced:')
df[df['simplified_tumor']=='BREAST'].groupby('Sex').count()['Age_at_diagnosis']
```

```python
plt.figure(figsize=(6,3))
sns.barplot(y='simplified_tumor', x= 'Age_at_diagnosis', hue='Sex',palette=['r','b'], data=df)

# Adjust figure
plt.tight_layout()
```

__Conclusion (2)__
* Male seems to have a later age at diagnosis than female in each tumor site, which seems to contradict to the __conclusion (1)__ 
* The conclusion (1) might be affect by the __unbalanced size__ in __breast cancer__ because:  
&nbsp;&nbsp;&nbsp;(a) Breast cancer has later age of diagnosis   
&nbsp;&nbsp;&nbsp;(b) There are much more female (n=585) than male (n=2) patients in breast cancer

## 3.  The effect of family case

#### 3.1. Barplot of the number of pateints by family case

```python
family_cases = ['proband','1st degree relative','2nd degree relative','3rd degree relative']

plt.figure(figsize=(6,2))
ax = sns.countplot(y='FamilyCase_group',data=df[df['FamilyCase_group'].isin(family_cases)])
```

#### 3.2. Barplot of the age at diagnosis by family case

```python
plt.figure(figsize=(6,2))
ax = sns.barplot(y='FamilyCase_group', x= 'Age_at_diagnosis', data=df[df['FamilyCase_group'].isin(family_cases)])
```

```python
# examine whether the distribution of age at diagnosis is Gaussian distribution
plt.figure(figsize=(12, 8))
for i, family_case in enumerate(family_cases):
    age_disr = df[df['FamilyCase_group']==family_case]['Age_at_diagnosis']
    plt.subplot(2,2,i+1)
    age_disr.plot.hist(bins=20,title=family_case)
    _, p = stats.normaltest(age_disr)
    if p < 0.05:
        print("{} is not Gaussian distribution (the null hypothesis can be rejected, p={:.1e})".format(family_case, p))
    else:
        print("{} is Gaussian distribution (the null hypothesis cannot be rejected, p={:.1e})".format(family_case, p))
```

#### 3.3. Compare the 'age at diagnosis' difference with Mann-Whitney test 

```python
cal_age_diff(family_cases,'FamilyCase_group')
```

#### 3.4. Barplot of the age at diagnosis by family case in each cacner
Since the tumor type is strongly associated with age of diagnosis, I took tumor type into consideration

```python
plt.figure(figsize=(6,4))
ax = sns.countplot(y='simplified_tumor', hue='FamilyCase_group', data=df[df['FamilyCase_group'].isin(family_cases)])
```

* Most of adrenal gland patients are proband, this might because of early onset for this tumor type.

```python
plt.figure(figsize=(6,4))
ax = sns.barplot(y='simplified_tumor', x= 'Age_at_diagnosis', hue='FamilyCase_group',
                 data=df[df['FamilyCase_group'].isin(family_cases)])
```

* Proband still has a ealier age of onset than other family classes in major cancer types

__Conclusion (3)__
* Proband has a significantly earlier age of onset than other family classes overall and in specific cancers
